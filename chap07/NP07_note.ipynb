### 패키지

하나의 모듈로 처리하기 어려울 때, 패키지를 사용

폴더에 \_\_init__.py 파일(초기화)이 있다면, 해당폴더를 패키지로 인식

모듈은  
import 함수명 으로 사용하지만
패키지는  
from 패키지명 import 모듈명 으로 사용

### 파이썬 가상환경

파이썬이 업데이트 되면서 옛날 코드가 작동하지 않을 수 있음!

이때, 가상환경을 사용하면 프로젝트 별 다른 버전 사용 가능

만드는 방법?
- venv : 파이썬 3.3 이상 내장 모듈
- conda : anaconda 설치 후 사용가능
- virtualenv : 써드 파티 도구(모듈 설치 후 가능)

프로젝트 시, 팀장이 모듈 등의 버전을 전부 정하고 팀원에게 넘길 때?

pip freeze > requirements.txt 로 저장

이후 해당 파일만 전달

받으면  
pip install -r requirements.txt 로 설치
